```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
 
% matplotlib inline
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sin value")
plt.title("Sine Wave")
plt.show()
```

```python
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.title("parabola")
plt.show()
```

```python
plt.plot(x,y, 'r+')
```

```python
plt.plot(x,y,'g.--')
```

```python
plt.plot(x,y,'yx')
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x), 'g-')
plt.plot(x,np.cos(x),'r.')
plt.xlabel("Angle")
plt.ylabel("Sine Wave")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['c','c++','java','kotlin','python','c#','php']
proficiency = [80,89,90,86,93,80,60]
ax.bar(langs,proficiency)
plt.xlabel('Languages')
plt.ylabel('Proficiency %')
plt.title("Bar Chart")
plt.show()
```

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
langs = ['c','c++','java','kotlin','python','c#','php']
proficiency = [80,89,90,86,93,80,60]

ax.pie(proficiency,labels=langs,autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,c='r')
ax.scatter(x,z,c='b')
```

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.title("3d projections")
```
